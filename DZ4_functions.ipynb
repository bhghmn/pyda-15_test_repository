```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

def main():
  while True:
    user_input = input("Введите команду: ")
    if user_input not in ['q', 'p', 's', 'i', 'ad', 'ds']:
      print('Команда не описана')
    if user_input == 'p':
      p()
    if user_input == 's':
      s()
    if user_input == 'i':
      i() 
    if user_input == 'das':
      ad() 
    if user_input == 'ds':
      ds()
    if user_input == 'q':
      print('До свидания!')
      break


def p(numbers = documents):
    user_input1 = input('Введите номер документа: ')
    for element in numbers:
        if element['number'] == user_input1:
            print(f"Владелец документа: {element['name']} ")
            break
    else:  
        print('Документ не найден в базе.')
        

def s(polka = directories):
    user_input2 = input('Введите номер документа: ')
    for key, value in polka.items():
        if user_input2 in value:
            print(f'Документ хранится на полке: {key}')                    
            break          
    else:  
        print('Документ не найден в базе.')
        
def i(polka = directories):
    for num, doc in enumerate(documents):
        for key, value in directories.items():
            if doc['number'] in value:
                print(f"документ: {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {key}")
                break 
  
def ad (polka = directories):
    user_input3 = input('Введите номер полки: ')
    if user_input3 in directories.keys():
        print(f"Такая полка уже существует. Текущий перечень полок: {list(directories.keys())}")
    else:
        polka[user_input3] = 34
        print(f"Полка добавлена. Текущий перечень полок: {list(directories.keys())}")
      

def ds(polka = directories):
    user_input4 = input('Введите номер полки: ')
    if user_input4 in directories.keys() and directories[user_input4] == []:
        del(polka[user_input4])
        print(f"Полка удалена. Текущий перечень полок: {list(directories.keys())}")
    elif user_input4 not in directories.keys():
        print(f"Такой полки не существует. Текущий перечень полок: {list(directories.keys())}")    
    else:   
        print(f"На полке есть документ, удалите его. Текущий перечень полок: {list(directories.keys())}")        
         
        
main()      
 
```
