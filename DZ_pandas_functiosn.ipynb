```python
# Задание 1
ratings = pd.read_csv('ratings.csv')
ratings.head(7)
```

```python
def rang(param):
    if  param <= 2:
        return 'низкий рейтинг'
    elif param <= 4:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'

ratings['class'] = ratings['rating'].apply(rang)
ratings.head(10)
```

```python
keywords = pd.read_csv('keywords.csv')
keywords.head(567)
```

```python
""" 
Не получилось сказать - 'если хотя бы один из элементов keyword' (сейчас выводит значения для всех слов в строке) """

def subekt(param):
    for element in param.split(' '):
        for key, value in geo_data.items():
            if element in value:
                return key
        else:
            return 'undefined'
    
        
keywords['region'] = keywords['keyword'].apply(subekt)
keywords.head(567)

# subekt('погода в санкт петербурге')
```
