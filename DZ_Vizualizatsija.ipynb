```python
print('Значение моды для Healthy life expectancy: ', df_new['Healthy life expectancy'].round(1).mode()[0])
```

```python
"""
Построим распределение ожидаемой продолжительности здоровой жизни. Из боксплота видим, что к выбросу 
отннесены данные по Короле́вству Эсвати́ни, для которого Healthy life expectancy в датасете определен как 0,0.
Чаще всего Healthy life expectancy лежит в дипозоне от 0,5 до 0,9, чаще всего принимает значение около 0,8
"""
df_new[['Healthy life expectancy']].plot(kind='box',
                                title='Distribution of odds')
```

```python
# посмотрим на взимосвязь показателя Generosity от GDP per capita
"""
Из получееной визуализации я бы выдвинула гипотезу, что щедрость жителей, не определяется уровнем
материального благосостояния страны, в которой они проживают.
"""
# посмотрим на взимосвязь показателей Healthy life expectancy, Social support от GDP per capita
"""
Из получееной визуализации я бы выдвинула гипотезу, что ожидаемая продолжительность здоровой жизни жителей, 
как и возможность их социальной поддержки имеют зависимость от материального благосостояния страны, 
в которой они проживают.
"""

df_new.plot(kind='scatter', 
        x='GDP per capita', 
        y='Generosity', 
        title='Correlation between Generosity and GDP per capita')
plt.show()
df_new.plot(kind='scatter', 
        x='GDP per capita', 
        y='Healthy life expectancy', 
        title='Correlation between Healthy life expectancy and GDP per capita')
plt.show()
df_new.plot(kind='scatter', 
        x='GDP per capita', 
        y='Social support', 
        title='Correlation between Social support and GDP per capita')
```

```python
# Посмотрим распределение параметра Freedom to make life choices, используя не boxplot, а distplot
ax = sns.distplot(df_new['Healthy life expectancy'])
ax.set_title('Healthy life expectancy distribution')
```
